```python
#loading file

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
data = pd.read_csv("/content/sample_data/Students_habits copy.csv")

# Display first 5 rows
data.head()

#checking and removing null values
data.isnull().sum()
null=data.dropna()
null.isnull().sum()
```

```python
# Selecting numerical columns
numeric_df = data.select_dtypes(include=["int64", "float64"])

# Correlation matrix
corr_matrix = numeric_df.corr()

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
#cap unrealistic values
for index, row in data.iterrows():
  if row['study_hours_per_day'] > 24:
    data.drop(index, inplace=True)
data.info()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data["exam_score"], bins=20, kde=True)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="study_hours_per_day", y="exam_score", data=data)
plt.title("Study Hours per Day vs Exam Score")
plt.xlabel("Study Hours per Day")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x="social_media_hours", y="exam_score", data=data)
plt.title("Social Media Hours vs Exam Score")
plt.xlabel("Social Media Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x="netflix_hours", y="exam_score", data=data)
plt.title("Netflix Hours vs Exam Score")
plt.xlabel("Netflix Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x="attendance_percentage", y="exam_score", data=data)
plt.title("Attendance Percentage vs Exam Score")
plt.xlabel("Attendance Percentage")
plt.ylabel("Exam Score")
plt.show()
```

Group Comparisons:
